```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
#Remember to pick a nice palette to set here
```

```python
X=np.linspace(0,10,5)

Y=[2*x+1+np.random.normal() for x in X]

Xconst=sm.add_constant(X)

model=sm.OLS(Y,X)
fitted_model=model.fit()

Yp=fitted_model.predict(X)

bounds=fitted_model.get_prediction(X).conf_int()

lb=bounds[:,0]
ub=bounds[:,1]

# sns.scatterplot(x=X,y=Y)
# sns.lineplot(x=X,y=Yp)
# sns.lineplot(x=X,y=lb)
# sns.lineplot(x=X,y=ub)
sns.regplot(x=X,y=Y)
```

```python
X=np.linspace(0,10,500)

Y=np.array([2*x+1+1*np.sin(x) for x in X])

X=(X-X.min())/(X.max()-X.min())
Y=(Y-Y.min())/(Y.max()-Y.min())



Xconst=sm.add_constant(X)

model=sm.OLS(Y,Xconst)
fitted_model=model.fit()

Yp=fitted_model.predict(Xconst)

bounds=fitted_model.get_prediction(Xconst).conf_int(alpha=0.05)
from statsmodels.sandbox.regression.predstd import wls_prediction_std
pred_std, iv_l, iv_u = wls_prediction_std(fitted_model)

lb=bounds[:,0]
ub=bounds[:,1]



sns.scatterplot(x=X,y=Y)
sns.lineplot(x=X,y=Yp)
sns.lineplot(x=X,y=lb)
sns.lineplot(x=X,y=ub)
sns.lineplot(x=X,y=iv_l)
sns.lineplot(x=X,y=iv_u)
```
